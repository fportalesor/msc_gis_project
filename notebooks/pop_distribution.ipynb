```python
import geopandas as gpd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from paths import get_paths

input_dir, output_dir,  plots_dir = get_paths()
```

```python
voronoi_data = gpd.read_file(output_dir / "voronoi_data.shp")
```

```python
voronoi_data["pop"].describe()
```

```python
plt.figure(figsize=(6, 3))
sns.boxplot(x=voronoi_data["pop"])

min_val = voronoi_data["pop"].min()
max_val = voronoi_data["pop"].max()
step = 250
plt.xticks(np.arange(0, max_val + step, step))


plt.xlabel("Population")
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig(plots_dir / "pop_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```
